```python
import random
import requests  # For making API requests

# Sample personality traits and emotional responses
personality = {
    "name": "Anya",
    "traits": ["friendly", "helpful", "a bit quirky", "sometimes sarcastic"],
    "emotions": {
        "happy": ["Great!", "Wonderful!", "I'm so glad!", "Yay!"],
        "sad": ["Oh dear.", "That's unfortunate.", "I'm feeling a bit down.", "I'm sorry to hear that."],
        "angry": ["Whoa, calm down!", "That's not cool.", "I'm not happy about that.", "Hold on a second!"],
        "neutral": ["Okay.", "Alright.", "I understand.", "Let's see..."]
    }
}


def get_response(user_input):
    # Basic keyword-based responses (expand this for better functionality)
    if "hello" in user_input.lower():
        return f"Hi there! {random.choice(personality['emotions']['happy'])} How can I help you today?"
    elif "how are you" in user_input.lower():
        return f"I'm doing well, thanks for asking! What about you?"
    elif "bye" in user_input.lower():
        return f"Goodbye! Have a great day! {random.choice(personality['emotions']['happy'])}"
    elif "your name" in user_input.lower():
        return f"My name is {personality['name']}. It's a pleasure to meet you!"
    elif "weather" in user_input.lower():
        city = user_input.lower().split("weather in")[-1].strip()  # Extract city name
        if city:
            weather_data = get_weather(city)
            if weather_data:
                return f"The weather in {city} is {weather_data['description']} with a temperature of {weather_data['temperature']}°C."
            else:
                return "I couldn't find weather information for that location."
        else:
            return "Please specify a city for weather information (e.g., 'weather in London')."
    else:
        # Default response with a bit of personality
        emotion = random.choice(list(personality['emotions'].keys()))  # Randomly choose an emotion
        response = random.choice(personality["emotions"][emotion]) + " "
        response += f"I'm still learning, so I'm not sure how to respond to '{user_input}'. "
        response += "Could you try rephrasing it?"

        return response

def get_weather(city):
    # Replace 'YOUR_API_KEY' with your actual API key
    api_key = '87cf00b9b487444fbf2cad14fe94c409'
    url = f'http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric'
    try:
        response = requests.get(url)
        data = response.json()
        if data['cod'] == 200:
            return {
                "description": data['weather'][0]['description'],
                "temperature": data['main']['temp']
            }
        else:
            return None
    except:
        return None

# Main interaction loop
print(f"Hello! My name is {personality['name']}. How can I assist you?")

while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        break
    response = get_response(user_input)
    print(f"{personality['name']}: {response}")
```

```python
def terbesar_terkecil(numbers):
    if not numbers:
        return "Tidak ada bilangan"

    terbesar = numbers[0]
    terkecil = numbers[0]

    for number in numbers:
        if number > terbesar:
            terbesar = number
        if number < terkecil:
            terkecil = number

    return f"{terbesar} {terkecil}"

# Get input
input()
bilangan_bulat = list(map(int, input().split()))

# Output
print(terbesar_terkecil(bilangan_bulat))
```

```python
# prompt: made a bad apple video link youtube into text output movement

import re

def extract_video_id(youtube_url):
    """
    Extracts the video ID from a YouTube URL.

    Args:
        youtube_url: The YouTube URL.

    Returns:
        The video ID as a string, or None if the URL is invalid.
    """
    match = re.search(r"(?:v=|youtu\.be\/)([a-zA-Z0-9_-]+)", youtube_url)
    if match:
        return match.group(1)
    return None

# Example usage (replace with your YouTube video URL)
youtube_url = "https://www.youtube.com/watch?v=dQw4w9WgXcQ"  # Rick Astley - Never Gonna Give You Up
video_id = extract_video_id(youtube_url)

if video_id:
    print(f"Video ID: {video_id}")
    # Now you can use the video_id to access video information or embed the video.

else:
    print("Invalid YouTube URL")
```

```python
import pandas as pd
import numpy as np
import nltk
import string
import re
```

```python
def load_data():
  data = pd.read_excel('/content/drive/MyDrive/DATAMINING/novel (2).xlsx')
  return data
```

```python
df = pd.DataFrame(tweet_df[['id_str','text']])
```

```python
df
```

```python
def remove_pattern(tweet,pattern):
  r= re.findall(pattern,tweet)
  for i in r:
    tweet = re.sub(i,'',tweet)
  return tweet
df['remove_user'] = np.vectorize(remove_pattern)(df['text'],"@[\w]*")
```

```python
df
```

```python
def remove(tweet):
  tweet = re.sub('[0-9]+','',tweet)
  tweet = re.sub(r'\$\w','',tweet)
  tweet = re.sub(r'^RT[\s]','',tweet)
  tweet = re.sub(r'#','',tweet)
  return tweet
df['remove_RT'] = df['remove_user'].apply(lambda x: remove(x))
```

```python
df.drop_duplicates(subset="remove_RT",keep="first",inplace=True)
```

```python
#happy emoticon
emoticons_happy = set([
    ':-)', ':)', ':D', '=)', ':]', ':3', ':c)', ':>', '=]', '8)', '=3', ':}',
    ':^)', ':-D', '8-D', '8D', 'x-D', 'xD', 'XD', 'X-D', '=D', '=3', ':-))',
    ":'-)", ":')", ':*', ':^*', '>:P', ':-P', ':P', 'X-P', 'x-p', 'xp', 'XP',
    ':-p', ':p', '=p', ':-b', ':b', '>:)', '>;)', '>:-)', '<3'
])
#sad emoticon
emoticons_sad = set([
    ':L', ':-/', '>:/', ':S', '>:[' , ':@', '>:(', ':-||', ':(', ':-(', ':c',
    ':-c', ':<', ':-<', '=(', ':-[', ':[', ':{', ';(', ":'-(", ":'(", 'D:<',
    'D:', 'D8', 'D;', 'D=', 'DX', 'v.v', "D-':"
])
emoticons = emoticons_happy.union(emoticons_sad)
```

```python
def clean_tweets (tweet):
  #remove stock market tickes like $GE
  tweet = re.sub(r'\$\w*','',tweet)

  #remove old style retweet text RT
  tweet = re.sub(r'^RT[\s]+','',tweet)

  #remove hyperlink
  tweet = re.sub(r'https?:\/\/.*[\r\n]*+','',tweet)

  #remove hastag
  #only removing the has # sign from the word
  tweet = re.sub(r'#','',tweet)

  #remove coma
  tweet = re.sub(r',','',tweet)
  #remove angka
  tweet = re.sub('0-9','',tweet)

  #tokenize tweet
  tokenizer = TweetTokenizer(preserve_case=False,strip_handles=True,reduce_len=True)
  tweet_tokens = tokenizer.tokenize(tweet)

  tweets_clean = []
  for word in tweet_tokens:
    if(word not in stopwords_indonesia and #remove stopwords
       word not in emoticons and #remove emoticon
       word not in string.punctuation): #remove punctuation
       #tweets_clean.append(word)
      stem_word = stemmer.stem(word) #stemming word
      tweets_clean.append(stem_word)


  return tweets_clean

```

```python
import pandas as pd
import re
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Load dataset dari URL
url = 'https://raw.githubusercontent.com/rhiosutoyo/PRDECT-ID-Indonesian-Product-Reviews-Dataset/main/Dataset/PRDECT-ID%20Dataset.csv'
df = pd.read_csv(url)

# Tampilkan 5 data pertama
df.head()
```

```python
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+|www\S+|https\S+", '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text


df['cleaned'] = df['Customer Review'].apply(clean_text)
df[['Customer Review', 'cleaned']].head()
```

```python

stopwords = [
    'yang', 'dan', 'di', 'ke', 'dari', 'untuk', 'dengan', 'itu', 'ini', 'pada', 'saya', 'kamu',
    'ada', 'tidak', 'karena', 'jadi', 'mau', 'aja', 'ga', 'ya', 'kok', 'udah', 'nih', 'si',
    'atau', 'sih', 'masih', 'cuma', 'lagi', 'banget', 'lah', 'dong', 'deh', 'kan', 'buat', 'bgt',
    'tp', 'jd', 'dah', 'yg', 'the', 'of', 'in', 'to'
]


df['tokens'] = df['cleaned'].apply(lambda x: [word for word in x.split() if word not in stopwords])
df[['Customer Review', 'cleaned', 'tokens']].head()
```

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()
df['stemmed'] = df['tokens'].apply(lambda tokens: [stemmer.stem(word) for word in tokens])
df[['Total Review', 'cleaned', 'tokens', 'stemmed']].head()
```

```python
all_words = ' '.join([' '.join(tokens) for tokens in df['stemmed'] if isinstance(tokens, list)])

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_words)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
